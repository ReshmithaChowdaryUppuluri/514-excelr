```python
import nltk
import string
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download("punkt_tab")
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
  """
  This function preprocesses a single text string.

  Args:
    text: The input text string.

  Returns:
    The preprocessed text string.
  """
  tokenize = word_tokenize(text)
  tokenize = [word.lower() for word in tokenize]
  stop_words = set(stopwords.words("english"))
  tokenize = [word for word in tokenize if word not in stop_words]
  lemmatizer = WordNetLemmatizer()
  tokenize = [lemmatizer.lemmatize(word) for word in tokenize]
  return " ".join(tokenize)

# Assuming 'df' is your DataFrame
df['text1_clean'] = df['text'].astype(str).apply(preprocess_text)
df['text2_clean'] = df['selected_text'].astype(str).apply(preprocess_text)
df.head()
```

```python
vectorizer=TfidfVectorizer()
Tfidf_matrix1=vectorizer.fit_transform(df['text1_clean'])
Tfidf_matrix2=vectorizer.fit_transform(df['text2_clean'])
```

```python
def jacob_similarity(review1, review2):
  w1 = set(review1.split())
  w2 = set(review2.split())
  intersection = len(w1.intersection(w2))
  union = len(w1.union(w2))
  # Handle the case where the union is empty to avoid ZeroDivisionError
  if union == 0:
    return 0  # Or any other appropriate value, like np.nan
  else:
    jacob_similarity = intersection / union
  return jacob_similarity
```

```python
df['jacob_similarity']=df.apply(lambda x: jacob_similarity(x['text1_clean'],x['text2_clean']),axis=1)
df.head()
```
